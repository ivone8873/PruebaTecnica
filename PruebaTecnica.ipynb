Obtener df

```python
# obtener datos
import pandas as pd
import os
import numpy as np
from pycaret.classification import *

df=pd.read_csv(os.path.join(os.getcwd(),"default_dataset.csv"))
df.columns
```

```python
df.shape
```

Estadísticas Descriptivas

```python
df.describe()
```

Pasamos ID a index

```python
df.set_index('ID',inplace=True)
```

Estandarización de los datos

```python
df_norm=(df-df.min())/(df.max()-df.min())
df_norm.describe()
```

Matriz de Correlación 

```python
df_norm.corr()
```

Se eliminan variables

```python
columns=['LIMIT_BAL','PAY_0', 'PAY_2','BILL_AMT1', 'BILL_AMT2',
       'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
       'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',  'PAY_AMT1',
       'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6', 'default']
df_norm=df_norm[columns]
```

```python
df_norm.columns
```

Se divide DF estandarizado 

```python

df_1 = df_norm.sample(frac=0.95, random_state=786)
```

```python
df_unseen = df_norm.drop(df_1.index)
df_unseen
```

Se reestablece el índice en los DF nuevos


```python
df_1.reset_index(inplace=True, drop=True)
df_unseen.reset_index(inplace=True, drop=True)
df_1.reset_index()
df_unseen.reset_index()
print('Data for Modeling: ' + str(df_1.shape))
print('Unseen Data For Predictions: ' + str(df_unseen.shape))
```

Configuración de PyCaret

```python
model_setup = setup(data=df_1, target='default', session_id=123)
```

Comparación de modelos

```python
best_model = compare_models()
```

Mejor Modelo

```python
print(best_model)
```

Lista de modelos

```python
models()
```

Creación del Modelo

```python
gbc = create_model('gbc')
```

```python
print('gbc')
```

Ajuste de hiperparámetros

```python
tuned_gbc = tune_model(gbc)
```

```python
print(tuned_gbc)
```

Gráficas del modelo

```python
# AUC
plot_model(tuned_gbc, plot = 'auc')
```

```python
# PRECISION RECALL
plot_model(tuned_gbc, plot = 'pr')
```

```python
# FEATURE IMPORTANCE 
plot_model(tuned_gbc, plot='feature')
```

```python
# CONFUSION MATRIX
plot_model(tuned_gbc, plot = 'confusion_matrix')
```

Evaluación del modelo

```python
evaluate_model(tuned_gbc)
```

Finalización del Modelo

```python
final_gbc = finalize_model(tuned_gbc)
```

```python
print(final_gbc)
```

Predicción del modelo

```python
predict_model(final_gbc)
```

```python
unseen_predictions = predict_model(final_gbc, data=df_unseen)
unseen_predictions
```
